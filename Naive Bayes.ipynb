```python
bow=array([[1,"amazing"],[2,'boy'],[3,'burn'],[4,'calculate'],[5,'expense'],[6,'fight'],[7,'gain'],[8,'game'],[9,'home'],[10,'healthy'],[11,'in'],[12,'joy'],[13,'know']])
print bow,'\n',shape(bow),'\n'
print bow[0][1]
           
```

```python
st=array([['i am seeing healthy fight in game'],['i know you calculate wrong expense']])
print st,'\n',shape(st),'\n',st[1][0]
```

```python
def feature_vector(bow, obj):
    n=shape(bow)[0]
    l=shape(obj)[0]
    vector=np.empty([l,n])
    for i in range(0,l):
        for j in range(0,n):
            if(bow[j][1] in obj[i][0]):
                vector[i][j]=1
            else:
                vector[i][j]=0
    return vector    
```

```python
obj=feature_vector(bow,st)
print obj
```
